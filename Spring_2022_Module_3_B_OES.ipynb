Just the basics. You need clean data that you 
want to check for correlations.

```r
### Code block 1
# load your data
FICO = read.csv(url("https://raw.githubusercontent.com/matsonah/ClarkeStatsSpring2022/main/data/Fico%20Avg%20and%20Top.csv"))
```

```r
### Code block 2
# specify your data & your desired values
data_x = FICO$FicoAvg 
data_y = FICO$FicoTop
desired_x = c(700, 683)
```

```r
### Code block 3
# use R to make a prediction.

data_model = lm(data_y ~ data_x)
predictions = data.frame(
                desired_x, 
                predicted_y=data_model$coefficients[2] * desired_x + data_model$coefficients[1]  
                )

xlb = min(data_x, desired_x)
xub = max(data_x, desired_x) 

ylb = min(data_y, predictions$predicted_y)
yub = max(data_y, predictions$predicted_y)

model_corr = cor(data_x, data_y)

predictions

plot(data_x, data_y, xlim=c(xlb,xub), ylim=c(ylb, yub))
points(predictions, cex=3, col="dark green")

cat(sep="", "The correlation coefficient is ", model_corr ,".\n")
```
