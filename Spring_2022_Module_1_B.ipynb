```r
### Code Block 4 
# Requires a green check mark next to Code Block 1.
# This code block pulls data completely at random. 
# Imagine rolling a die every time to decide which 
# observation to pull.

sample_size = 30      
srs.sample <- fulldata[ 
                  sample( 1:nrow(fulldata),
                          sample_size,
                          replace = FALSE),
                      ]

# What does srs.sample stand for? (Think of the vocab list.)
# What happens to the results when you click play multiple times?

table(srs.sample$search_conducted) # Pick different columns to personalize this code.
pie(table(srs.sample$search_conducted))
```

```r
### Code Block 5 
# Requires a green check mark next to Code Block 1.
# This code block pulls data systematically at random. 
# Imagine rolling a die for a starting point and then proceeding 
# every so often to pull a new observation. 


# defines the function "obtain_sys"
obtain_sys = function(N,n){
  k = ceiling(N/n)          # step size between data
  r = sample(1:k, 1)        # random starting point
  seq(r, r + k*(n-1), k)    # sequence of datum locations
}

# calls the function "obtain_sys" 
sys.sample = fulldata[obtain_sys(nrow(fulldata), 30), ]

# What does sys.sample stand for? (Think of the vocab list.)
# What happens to the results when you click play multiple times?

table(sys.sample$search_conducted)   # Pick different columns to personalize this code.
pie(table(sys.sample$search_conducted))
```
