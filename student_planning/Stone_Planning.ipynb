```r
week = 5
rate = 48/16
progress = rate * week
progress
```

```r
even_claims = c(rep(3,16))    # I'm too lazy to type sixteen threes.
weeks = seq(1,16,1)           # I'm too lazy to type all the numbers from 1 to 16. 
planned_claims = c(0,1,4,3,4,5,3,5,3,8,1,0,0,2,3,3)

par(mfrow=c(2,2))

plot(weeks, planned_claims, col="dark green", type="l") 
points(even_claims, col="blue", type="l")
points(x=week,y=rate, col="dark violet", pch=8, cex=3)

plot(weeks, cumsum(planned_claims), col="dark green", type ="l") 
points(cumsum(even_claims), col="blue", type="l")
points(x=week,y=progress, col="dark violet", pch=8, cex=3)
```
